```python
import skimage.io as ski
import skimage.transform as skt
```

```python
img = ski.imread('dataset_1/user_10_A1.jpg')
ski.imshow(img)
```

```python
from skimage.transform import SimilarityTransform
from skimage.transform import warp
tform = SimilarityTransform(translation=(0, -10))
warped = warp(img, tform)
```

```python
ski.imshow(warped)
```

```python
from skimage.transform import rotate
rotated = rotate(img,10,resize=True)
ski.imshow(rotated)
```

```python
from scipy import ndimage
from skimage import exposure

horizontal_flip = img[:, ::-1]
original_image = horizontal_flip
adjusted_gamma_image = exposure.adjust_gamma(original_image, gamma=0.4, gain=0.9)
transformed_image = blured_image = ndimage.uniform_filter(adjusted_gamma_image, size=(6,6, 1))

ski.imshow(transformed_image)
ski.imsave('transformed_image.jpg',transformed_image)
```

```python
from scipy import ndimage
from skimage import exposure
from scipy.misc import imfilter

horizontal_flip = img[:, ::-1]
transformed_image  = ndimage.uniform_filter(horizontal_flip, size=(1,1, 1))
ski.imshow(imfilter(transformed_image,'sharpen'))
```

```python
horizontal_flip = img[:, ::-1]
original_image = horizontal_flip
adjusted_gamma_image = exposure.adjust_gamma(original_image, gamma=0.4, gain=0.9)
transformed_image = blured_image = ndimage.uniform_filter(adjusted_gamma_image, size=(6,6, 1))
img_to_show = imfilter(transformed_image,'sharpen')
img_to_show = imfilter(img_to_show,'sharpen')
img_to_show = imfilter(img_to_show,'sharpen')

ski.imshow(img_to_show)
```
