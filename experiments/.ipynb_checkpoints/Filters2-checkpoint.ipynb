```python
from skimage import data, io, feature, filters
image = data.coins()
```

```python
io.imshow(image)
```

```python
gauss_edges = filters.gaussian(image,sigma=4)
canny_edges = feature.canny(gauss_edges,sigma=1)
io.imshow(canny_edges)
```

```python
from skimage.morphology import disk
median_edges=filters.median(image,disk(5))
canny_edges = feature.canny(median_edges,sigma=2)
io.imshow(canny_edges)
```

```python
io.imshow(data.chelsea())

img = data.chelsea()[:,:,1]
io.imshow(filters.sobel(img))
```

```python
import cv2
import numpy as np
img = cv2.imread('data/image_1.jpg',0)
io.imshow(img)
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations=5)
io.imshow(erosion)
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.dilate(img,kernel,iterations=3)
io.imshow(erosion)
```

```python
img2 = cv2.imread('data/image_2.jpg',0)
io.imshow(img2)
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img2,kernel,iterations=2)
io.imshow(erosion)
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.dilate(img2,kernel,iterations=3)
io.imshow(erosion)
```

```python
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)
io.imshow(opening)
```

```python
gradient = cv2.morphologyEx(img2, cv2.MORPH_GRADIENT, kernel)
io.imshow(gradient)
```
