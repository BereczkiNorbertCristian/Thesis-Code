```python
from skimage import feature, data, io, filters
```

```python
image = data.coins()
```

```python
sobel_edges=filters.sobel(image)
io.imshow(sobel_edges)
```

```python
sobel_edges_v=filters.sobel_v(image)
io.imshow(sobel_edges_v)
```

```python
sobel_edges_h=filters.sobel_h(image)
io.imshow(sobel_edges_h)
```

```python
canny_edges_1 = feature.canny(image,sigma=1)
io.imshow(canny_edges_1)
```

```python
canny_edges_1 = feature.canny(image,sigma=3)
io.imshow(canny_edges_1)
```

```python
laplace_edges=filters.laplace(image)
io.imshow(laplace_edges)
```

```python
edges=filters.roberts(image)
io.imshow(edges)
```

```python
edges=filters.prewitt(image)
io.imshow(edges)
```

```python
from skimage.morphology import disk
edges=filters.median(image,disk(5))
io.imshow(edges)
```

```python
edges=filters.gaussian(image,sigma=3)
io.imshow_collection(edges)
```

```python
from skimage.filters.rank import mean_bilateral
bilat_img = mean_bilateral(image, disk(20), s0=10,s1=10)
io.imshow(bilat_img)
```

```python
from skimage.filters.rank import subtract_mean
out = subtract_mean(image, disk(5))
io.imshow(out)
```

```python
from skimage.filters.rank import modal
out = modal(image, disk(5))
io.imshow(out)
```
